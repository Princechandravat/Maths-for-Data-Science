```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
np.random.seed(42)

heights=np.random.normal(loc=170,scale=10,size=200)

mean_val=np.mean(heights)
std_val=np.std(heights,ddof=1)
print(f"Sample mean: {mean_val:.2f} cm, Sample std dev: {std_val:.2f} cm")

fig = plt.figure()
ax = fig.add_subplot(111)
stats.probplot(heights, dist="norm", plot=ax)
ax.set_title("QQ Plot: Student Heights vs Theoretical Normal Distribution")
plt.show()
```

```python
# Continuous Uniform generate
data = np.random.uniform(0, 10, 1000)  # 0 se 10 ke beech 1000 samples

# Plot histogram
plt.hist(data, bins=20, density=True)
plt.title('Uniform Distribution Example')
plt.grid(True)
plt.show()

# Mean aur Variance check
print("Mean:", np.mean(data))  # ~5
print("Variance:", np.var(data))  # ~(10-0)^2 / 12 = 8.33
```
