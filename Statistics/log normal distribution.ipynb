# Log Normal Distribution 

```python
import numpy as np
mu, sigma = 0, 0.25
data = np.random.lognormal(mean=mu, sigma=sigma, size=1000)

sns.kdeplot(data)
plt.grid(True)
plt.show()
```

```python
log_data = np.log(data)
sns.kdeplot(log_data)
plt.grid(True)
plt.show()
```
