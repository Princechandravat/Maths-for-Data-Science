# Hypothesis Testing -

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import norm
```

```python
# Given values
sample_mean = 78
mu0 = 75
s = 10
n = 20

# t statistic
t_stat = (sample_mean - mu0) / (s / np.sqrt(n))

# Degrees of freedom
df = n - 1

# p-value (two-tailed)
p_value = stats.t.sf(abs(t_stat), df=df) * 2

# Generate t distribution curve
x = np.linspace(-4, 4, 500)
y = stats.t.pdf(x, df)

# Critical values for 95% confidence (alpha = 0.05)
alpha = 0.05
t_crit = stats.t.ppf(1 - alpha/2, df)

# Plot
plt.figure(figsize=(8,5))
plt.plot(x, y, label=f"t-distribution (df={df})")

# Shade rejection regions
plt.fill_between(x, y, where=(x <= -t_crit), color='red', alpha=0.3, label="Rejection region")
plt.fill_between(x, y, where=(x >= t_crit), color='red', alpha=0.3)

# Shade acceptance region
plt.fill_between(x, y, where=(x > -t_crit) & (x < t_crit), color='green', alpha=0.2, label="Fail to Reject H0")

# Plot t-statistic line
plt.axvline(t_stat, color='blue', linestyle='--', label=f"t-stat = {t_stat:.2f}")

# Labels
plt.title("One-Sample t-Test Visualization")
plt.xlabel("t value")
plt.ylabel("Density")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

print("t-statistic:", t_stat)
print("p-value:", p_value)
print("Critical value:", t_crit)
```

# Z-Test-

## Question 1-

A company claims that the average weight of their sugar packets is 1 kg. A random sample of 100 packets has a mean weight of 0.98 kg with a standard deviation of 0.05 kg.

Test at the 5% significance level whether the companys claim is true.

```python
# Given data
sample_mean = 0.98
pop_mean = 1.00
pop_std = 0.05
n = 100
alpha = 0.05

# Step 1: Standard Error
SE = pop_std / np.sqrt(n)

# Step 2: Z-score
z = (sample_mean - pop_mean) / SE

# Step 3: Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Step 4: Decision
print("Z-score:", z)
print("p-value:", p_value)
if p_value < alpha:
    print("Reject H0: The mean is not equal to 1 kg.")
else:
    print("Fail to reject H0: Not enough evidence against the claim.")
```

## Question 2-

The average life of a certain type of battery is 500 hours. A sample of 64 batteries has an average life of 490 hours with a population standard deviation of 20 hours.

```python
mean = 490
n = 64
mu = 500
std = 20
alpha = 0.001

# standard error -
standard_error = std/np.sqrt(n)
print(f'standard error : {standard_error}')

# z-score-
z = (mean - mu) / standard_error
print(f'Z score : {z}')

# p Value-
p_value = 2*(1-norm.cdf(abs(z)))
print(f'P value: {p_value}')

# decision-
if p_value < alpha :
    print('H0: Reject null hypothesis')
else:
    print('H0: Accept null hypothesis')
```

## Question 3-

A university claims that the average score of its students is at least 70. A sample of 36 students had a mean score of 68 with a population standard deviation of 6.

At the 5% significance level, test whether the claim is valid.

```python
mean = 68
mu = 70
n = 36
alpha = 0.05
std = 6

# standard error-
SE = std/np.sqrt(n)
print(f'Standard error : {SE}')

# Z Score-
z_score = mean - mu / SE
print(f'Z_Score {z_score}')

# p value-
p_value = norm.cdf(z_score)
print(f'P Value: {p_value}')

# Critical value at 5% significance value-
critical_z = norm.ppf(alpha)
print(f'Critical Value: {critical_z}')

if z_score < critical_z :
    print('H0 : Reject null hypothesis')
else:
    print('H0 : Accept null hypothesis')
```

## Question 4-

A poll claims that 60% of voters support a new policy. In a random sample of 400 voters, 220 support the policy.

 Test at 5% significance level whether the claim is correct.

```python
p0 = 0.60
x = 220/400
n = 400
alpha = 0.05

# standard_error-
Se = np.sqrt(p0*(1-p0)/n)
print(f'Standard error : {Se}')

# Z stats-
z = (x-p0)/Se
print( f'Z Stats : {z}')

# two-tailed-
p_value = 2*(1-norm.cdf(abs(z)))
print(f'P Value :{p_value}')

#Decision-

if p_value < alpha :
    print('Reject Null hypothesis (H0)')
else:
    print('Accept Null hypothesis (H0)')
```

## Question 5-

A company claims that 80% of its deliveries arrive on time. Out of 300 deliveries, 228 arrived on time.

At 5% significance level, test the claim.

```python
mu = 0.80
mean = 228/300
n = 300
alpha = 0.05

# standard_error-
SE = np.sqrt(mu*(1-mu)/n)
print(f'Standard_error : {SE}')

# Z stats-
z = (mean-mu)/SE
print(f'Z score : {z}')

# P Value-
p_value = 2*(1-norm.cdf(abs(z)))
print(f'P value : {p_value}')

# Decision-
if p_value>alpha :
    print('Reject null hypothesis (H0)')
else:
    print('Accept null hypothesis (H1)')
```
