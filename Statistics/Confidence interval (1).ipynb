# Confidence Interval-

```python
import numpy as np
from scipy import stats

data = [165, 167, 164, 166]  # tera sample
mean = np.mean(data)
se = stats.sem(data)  # standard error
ci = stats.norm.interval(0.95, loc=mean, scale=se)  # 95% CI
print(f"Confidence Interval: {ci}")
```

```python
# A/B Testing for website interval-

# version A-
p_a=0.20
n_a=200
standard_error_a= np.sqrt(p_a*(1-p_a)/n_a)
ci_a= stats.norm.interval(0.95,loc=p_a,scale=standard_error_a)
print('standard_error of a:-', standard_error_a)
print(f'95% class interval for version A: {ci_a}')

# Version B-
p_b=0.24
n_b= 250
standard_error_b = np.sqrt(p_b*(1-p_b)/n_b)
ci_b= stats.norm.interval(0.95,loc=p_b,scale=standard_error_b)
print( 'standard error of b:-', standard_error_b)
print(f'95% class interval for version B: {ci_b}')
```

```python
# Customer satisfaction Survey-

# A part-
#Given values-
n=100 # sample data
std=1.2 #standard deviation
mean=7.5 # mean
confidence_level=0.95 #confidence level
z_score_a=1.96

# Standard error-
standard_error= std/np.sqrt(n)  #standard error
print('standard_error:-',standard_error)

#Confidence Interval-
confidence_interval= stats.norm.interval(confidence_level,loc=mean,scale=standard_error)  # Confidence interval
print('Confidence_Interval:-', confidence_interval)


# B Part-

# Given Data 
std=1.2  # standard deviation
moe=0.2
z_score_b= 1.96

#Required Samples -
n= (z_score_b*std/moe)**2
print('Required sample size:-', n)

# Plotting
plt.figure(figsize=(8, 5))
plt.axvline(mean, color='blue', linestyle='--', label=f'Mean = {mean}')
plt.axvline(confidence_interval[0], color='green', linestyle='--', label=f'Lower CI = {confidence_interval[0]:.2f}')
plt.axvline(confidence_interval[1], color='red', linestyle='--', label=f'Upper CI = {confidence_interval[1]:.2f}')

# Distribution curve for visualization
x = np.linspace(mean - 4*standard_error, mean + 4*standard_error, 1000)
y = stats.norm.pdf(x, mean, standard_error)
plt.plot(x, y, color='black')

plt.title("Confidence Interval for Customer Satisfaction Mean (95%)")
plt.xlabel("Satisfaction Score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Machine Learning Model Accuracy-

# Given data
p_hat = 0.82
n = 500
se = np.sqrt((p_hat * (1 - p_hat)) / n)  #progressivement Standard Error
print(f"Standard Error: {se:.5f}")

# 90% CI
ci_90 = stats.norm.interval(0.90, loc=p_hat, scale=se)
print(f"90% CI: [{ci_90[0]*100:.2f}%, {ci_90[1]*100:.2f}%]")

# 99% CI
ci_99 = stats.norm.interval(0.99, loc=p_hat, scale=se)
print(f"99% CI: [{ci_99[0]*100:.2f}%, {ci_99[1]*100:.2f}%]")

# Compare widths
width_90 = ci_90[1] - ci_90[0]
width_99 = ci_99[1] - ci_99[0]
print(f"Width of 90% CI: {width_90*100:.2f}%")
print(f"Width of 99% CI: {width_99*100:.2f}%")

# Visualization
confidence_levels = ['90% CI', '99% CI']
proportions = [p_hat, p_hat]
errors = [se * 1.645, se * 2.576]
plt.errorbar(confidence_levels, proportions, yerr=errors, fmt='o', capsize=5)
plt.ylabel('Accuracy')
plt.title('90% and 99% Confidence Intervals for Model Accuracy')
plt.ylim(0.7, 0.9)
plt.grid(True)
plt.show()
```

```python
# Practice question -

# A Testing-
point_estimate_A = 0.20
total_visitors_A = 200

#standard error-
standard_error_a = np.sqrt(point_estimate_A*(1-point_estimate_A)/total_visitors_A)
print(f'standard_error of test A = {standard_error_a}')
#confidence interval-
confidence_interval_A = stats.norm.interval(0.95,loc=point_estimate_A,scale=standard_error_a)
print(f'confidence_interval of A = {confidence_interval_A}')


# B Testing-
point_estimate_B = 0.24
total_visitors_B = 250

#standard error-
standard_error_B = np.sqrt(point_estimate_B*(1-point_estimate_B)/total_visitors_B)
print(f'standard error of Test B = {standard_error_B}')

#confidence interval-
confidence_interval_B = stats.norm.interval(0.95,loc=point_estimate_B,scale=standard_error_B)
print(f'confidence_interval of B = {confidence_interval_B}')


# Errors for error bars
error_A = confidence_interval_A[1] - point_estimate_A
error_B = confidence_interval_B[1] - point_estimate_B

# Visualization
estimates = [point_estimate_A, point_estimate_B]
errors = [error_A, error_B]
labels = ['Test A', 'Test B']

plt.errorbar(labels, estimates, yerr=errors, fmt='o', capsize=5, color='blue')
plt.ylabel('Conversion Rate')
plt.title('A/B Testing with 95% Confidence Intervals')
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
