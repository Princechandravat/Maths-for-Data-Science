```python
import numpy as np
import matplotlib.pyplot as plt

# Population (skewed distribution)
np.random.seed(42)
population = np.random.exponential(scale=2.0, size=100000)

# Function to take sample means
def sample_means(n, draws=5000):
    return [np.mean(np.random.choice(population, n)) for _ in range(draws)]

# Take sample means for n=5, 30, 100
means_5 = sample_means(5)
means_30 = sample_means(30)
means_100 = sample_means(100)

# Plot histogram
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.hist(means_5, bins=30, color="skyblue", edgecolor="black")
plt.title("Sample size = 5")

plt.subplot(1,3,2)
plt.hist(means_30, bins=30, color="orange", edgecolor="black")
plt.title("Sample size = 30")

plt.subplot(1,3,3)
plt.hist(means_100, bins=30, color="green", edgecolor="black")
plt.title("Sample size = 100")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_rolls = 1000       # number of experiments
sample_size = 30     # sample size per experiment

means = []

for _ in range(n_rolls):
    sample = np.random.randint(1, 7, sample_size)  # dice rolls
    means.append(np.mean(sample))

# Plot histogram of sample means
plt.hist(means, bins=30, density=True, color='skyblue', edgecolor='black')
plt.title("CLT Example: Dice Rolls (Sample Means)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.show()
```

```python
# Population: Exam marks (skewed distribution)
np.random.seed(42)
population = np.random.exponential(scale=20, size=100000)  # 100k students

plt.hist(population, bins=50, color='skyblue', edgecolor='black')
plt.title("Population Distribution (Exam Marks)")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
sample_size=30
experiment_time=1000

sample_mean=[]
for i in range(experiment_time):
    sample=np.random.choice(population,size=sample_size,replace=False)
    sample_mean.append(np.mean(sample))

mean=np.mean(sample_mean)
print('mean of sample_mean :-',mean)

sns.kdeplot(sample_mean)
plt.show()
```
