```python
from scipy.stats import binom

n = 5
p = 0.5

# P(X <= 2) = cdf(2)
prob_less_equal_2 = binom.cdf(2, n, p)

# Complement
prob_3_or_more = 1 - prob_less_equal_2

print("P(X >= 3) using complement rule:", prob_3_or_more)
```
