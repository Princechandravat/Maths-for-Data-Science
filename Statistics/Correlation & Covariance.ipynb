# Covariance and Correlation-

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
tips.info()
```

```python
data_corr=tips.select_dtypes(['float64','int64']).corr()
data_corr
```

```python
data_cov=tips.select_dtypes(['float64','int64']).cov()
data_cov
```

```python
#correlation-
plt.figure(figsize=(4,3))
sns.heatmap(data_corr,annot=True)
plt.show()
```

```python
#covariance-

plt.figure(figsize=(4,3))
sns.heatmap(data_cov,annot=True)
plt.show()
```

```python
#dataset-

salary=pd.read_csv(r"C:\Users\Prince Chandravat\OneDrive\Videos\Documents\DATASETS\employee salary dataset\synthetic_salary_dataset.csv")
salary.head()
```

```python
salary.describe()
```

```python
salary.info()
```

```python
#mean-
mean=salary['Salary(INR)'].mean()
print('mean of salary:', mean)

#std-
std=round(salary['Salary(INR)'].std())
print('standard deviation of salary:',std)

#variance-
Variance=salary['Salary(INR)'].var()
print('Variance of salary:',Variance)
```

```python
# PERCENTILE-
Q1=np.percentile(salary['Salary(INR)'],25)
Q2=np.percentile(salary['Salary(INR)'],50)
Q3=np.percentile(salary['Salary(INR)'],75)

#IQR-
iqr=Q3-Q1

# min/max-
minimum=Q1-(1.5*iqr)
maximum=Q3+(1.5*iqr)

# outliers-

outliers= [ x for x in salary['Salary(INR)'] if x < minimum or x> maximum]


print('Q1:',Q1)  #q1
print('Q2:',Q2)  #q2
print('Q3:',Q3)  #q3
print('IQ:R',iqr)  #iqr
print('minimum:',minimum)  #minimum
print('maximum:',maximum)   #maximum
print('outliers:',outliers)  #outliers
```

```python
#Correlation-
salary_corr = salary.select_dtypes(['float64', 'int64']).corr()
```

```python
#heatmap-

plt.figure(figsize=(4,3))
sns.heatmap(salary_corr,annot=True)
plt.show()
```
