```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import bernoulli
from scipy.stats import uniform
```

# Bernouli Distribution

```python
# Question 1: Coin Toss Simulation (fair coin, p=0.5)
p = 0.5
n = 1000

# Generate samples (0 = tail, 1 = head)
samples = bernoulli.rvs(p, size=n, random_state=42)

# Count heads (1) and tails (0)
unique, counts = np.unique(samples, return_counts=True)
results = dict(zip(unique, counts))

# Plot bar chart
plt.figure(figsize=(5,4))
plt.bar(results.keys(), results.values(), tick_label=["Tails (0)", "Heads (1)"], color=["skyblue","orange"], edgecolor="black")
plt.title(f"Bernoulli Simulation (p={p}, n={n}) - Coin Toss")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', linewidth=0.5)
plt.show()

# Print results
print(results, f"Heads proportion = {results.get(1,0)/n:.3f} (theoretical = {p})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# probability of success (p)
p = 0.7  

# Generate random samples
samples = bernoulli.rvs(p, size=1000)

# Mean and variance
mean, var,kurtosis,skewness = bernoulli.stats(p, moments='mvks')
print("Mean:", mean)
print("Variance:", var)
print("kurtosis:", kurtosis)
print("skewness:", skewness)

# Plot histogram
plt.hist(samples, bins=2, rwidth=0.8, color='skyblue')
plt.xticks([0, 1], ['Failure (0)', 'Success (1)'])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title(f"Bernoulli Distribution (p={p})")
plt.show()
```

```python
# Probability of Head (Success)
p = 0.4  

# Probability of Tail
print("Probability of Tail:", 1 - p)

# Generate 1000 tosses
tosses = bernoulli.rvs(p, size=1000)

# Count Heads and Tails
unique, counts = np.unique(tosses, return_counts=True)
result = dict(zip(unique, counts))
print("Simulation Result:", result)

# Plot histogram
plt.hist(tosses, bins=2, rwidth=0.8, color='teal')
plt.xticks([0, 1], ['Tail (0)', 'Head (1)'])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Coin Toss Simulation (Bernoulli, p=0.4)")
plt.show()
```
