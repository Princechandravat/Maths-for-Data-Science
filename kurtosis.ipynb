```python
import pandas as pd
from scipy.stats import kurtosis
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kurtosis as sm
import numpy as np


# Example feature
data = [10, 12, 12, 13, 12, 100, 12, 13, 12, 11]

data2 = [1, 2, 2, 3, 3, 4, 5, 100]

# Calculate kurtosis
kurt_val = kurtosis(data)
print("Kurtosis:", kurt_val)
print('Kutosis',kurtosis(data2))

if kurt_val > 3:
    print("Heavy tails - possible outliers")
elif kurt_val < 3:
    print("Light tails - less extreme values")
else:
    print("Normal distribution")
```

```python
import numpy as np
from scipy.stats import kurtosis, norm, t, uniform

# Generate data for different distributions
np.random.seed(42)  # For reproducibility

# 1. Mesokurtic: Normal distribution
normal_data = norm.rvs(size=1000)  # Mean 0, std 1
normal_kurt = kurtosis(normal_data, fisher=True)  # Fisher=True for excess kurtosis

# 2. Leptokurtic: Student's t-distribution (df=3, heavy tails)
lepto_data = t.rvs(df=3, size=1000)
lepto_kurt = kurtosis(lepto_data, fisher=True)

# 3. Platykurtic: Uniform distribution
platy_data = uniform.rvs(size=1000)
platy_kurt = kurtosis(platy_data, fisher=True)

# Print results
print("Normal (Mesokurtic) Excess Kurtosis:", normal_kurt)  # ~0
print("T-Dist (Leptokurtic) Excess Kurtosis:", lepto_kurt)  # >0, often around 6+
print("Uniform (Platykurtic) Excess Kurtosis:", platy_kurt)  # <0, around -1.2

# Custom data example
custom_data = np.array([1, 2, 3, 4, 5])
custom_kurt = kurtosis(custom_data, fisher=True)
print("Custom Data Excess Kurtosis:", custom_kurt)  # ~ -1.3 (platykurtic)
```

```python
import matplotlib.pyplot as plt

# Use the data from above
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Normal
axs[0].hist(normal_data, bins=30, density=True, alpha=0.6, color='g')
axs[0].set_title('Mesokurtic (Normal)')
axs[0].set_xlabel('Value')
axs[0].set_ylabel('Density')

# Leptokurtic
axs[1].hist(lepto_data, bins=30, density=True, alpha=0.6, color='r')
axs[1].set_title('Leptokurtic (T-Dist)')
axs[1].set_xlabel('Value')

# Platykurtic
axs[2].hist(platy_data, bins=30, density=True, alpha=0.6, color='b')
axs[2].set_title('Platykurtic (Uniform)')
axs[2].set_xlabel('Value')

plt.show()  # Ye window mein chart dikhaega
```
